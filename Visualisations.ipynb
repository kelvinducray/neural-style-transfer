```python
# from neural_style_transfer import image_handler, style_transfer, model
from PIL import Image
import matplotlib.pyplot as plt
from torchvision.models import vgg19

from neural_style_transfer import image_handler, style_transfer
```

```python
content_image = Image.open("./images/content/dancing.jpg")
style_image = Image.open("./images/style/picasso.jpg")

# assert (
#     style_img.size() == content_img.size()
# ), "we need to import style and content images of the same size"

image_handler.display_input_images(content_image, style_image)
```

# Transform images

```python
content_image_transformed = image_handler.transform_image(content_image)
style_image_transformed = image_handler.transform_image(style_image)
```

# Perform image style transfer

```python
style_transfer_config = style_transfer.StyleTransferConfig(
    model=vgg19(pretrained=True),
    feature_layers=[0, 5, 10, 19, 28],
)

style_transfer_model = style_transfer.StyleTransferModel(
    style_transfer_config=style_transfer_config,
)

style_transfer = style_transfer.StyleTransfer(
    content_image=content_image_transformed,
    style_image=style_image_transformed,
    style_transfer_model=style_transfer_model,
)
```

```python
style_transfer.optimise()
```
